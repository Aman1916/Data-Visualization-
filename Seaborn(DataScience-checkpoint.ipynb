# Bar plot in seaborn

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df2=sns.load_dataset("penguins")
df2.head(3)
```

```python
df2["island"]
```

```python
sns.barplot(x=df2.island,y=df2.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island",y='bill_length_mm',data = df2)
plt.show()
```

```python
#categorical = hue
sns.barplot(x = "island",y='bill_length_mm',data = df2,hue="sex")
plt.show()
```

```python
#order parameter

order1 = ['Dream',"Biscoe",'Torgersen']
sns.barplot(x = "island",y='bill_length_mm',data = df2,hue="sex",order=order1)
plt.show()
```

```python
order1 = ['Dream',"Biscoe",'Torgersen']
sns.barplot(x = "island",y='bill_length_mm',data = df2,hue="sex",order=order1,palette='gist_heat_r')
plt.show()
```

```python
order1 = ['Dream',"Biscoe",'Torgersen']
sns.barplot(x = "island",y='bill_length_mm',data = df2,hue="sex",order=order1,palette='gist_heat_r',saturation=0.1)
plt.show()
```
