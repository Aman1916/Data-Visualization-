#  Matplotlib

```python
import matplotlib.pyplot as plt
```

1. linear plot
2. Scatter plot
3. Bar plot
4. Stem plot
5. hist plot
6. Box plot

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
plt.plot(x,y,color="r")
plt.show()
```

```python
product_id=[1,2,3,4,5,6,7]
offline_price=[222,226,222,88,11,666,333]
online_price=[444,222,77,456,886,111,227]
chor_bazar=[323,654,345,223,123,421,112]
plt.plot(product_id,online_price,color='red',marker='*',markersize=10)
plt.plot(product_id,offline_price,color='green',marker='D',markersize=10)
plt.plot(product_id,chor_bazar,color='blue',marker='+',markersize=10)
plt.xlabel("product")
plt.ylabel("Prices")
plt.title("product price comprasion")
plt.show()
```

```python
x = [1,5,2,6,11,2,11,2]
y = [4,8,1,6,3,8,2,12]
plt.plot(x,y,color="r",marker = "*")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
c = ['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
x= np.array(["a","b","c","d"])
y= np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
x= np.array(["a","b","c","d"])
y= np.array([3,8,1,10])
plt.bar(x,y,color = "g",width=0.2)
plt.show()
```

# Scatter plot

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
plt.scatter(x,y,color="r")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
plt.title("sactter plot")
plt.xlabel("month")
plt.ylabel("year")
plt.scatter(x,y,color="g")
plt.show()
```

```python
x = [1,9,3,4,8,6,7,8,9]
y = [9,2,7,6,5,1,3,2,1]
plt.title("sactter plot")
plt.xlabel("month")
plt.ylabel("year")
c=['red','magneta','green','orange','blue','brown']
plt.scatter(x,y, color ="c")
plt.show()
```

# Matplotlib -Image usecase

```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fname=r'eye.jpg'
image = Image.open(fname).convert('L')
#mapping image to gray scale
plt.imshow(image,cmap ="gray")
plt.show()
```

```python
fname=r'eye.jpg'
image = Image.open(fname).convert('L')
#mapping image to gray scale
plt.imshow(image,cmap ='Blues')
plt.show()
```

```python
fname=r'eye.jpg'
image=Image.open(fname).convert('L')
plt.imshow(image,cmap="Spectral")
plt.show()
```
